[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## [1-1] 모듈로딩
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## [1-2] 데이터 준비
IMG_DIR = '../data/image/'
```

[2] 이미지 로딩 <hr>

```python
FILE = IMG_DIR+'strawberry.png'

strawberryImg = cv2.imread(FILE, cv2.IMREAD_COLOR)  ## cv2.IMREAD_COLOR_BAR와 동일
```

```python
## 이미지 정보 => ndarray
print(f'형태 : {strawberryImg.shape}')
print(f'차원 : {strawberryImg.ndim}D')
print(f'타입 : {strawberryImg.dtype}')
```

```python
## 이미지 출력
plt.imshow(strawberryImg)
plt.title('Strawberry')
plt.show()
```

```python
## 이미지 팝업 출력창 ==> 이미지 창 위에 클릭 후 키보드 키 아무거나 입력 ==> 창 닫김
cv2.imshow("[Strawberry]", strawberryImg) # 이미지 데이터 창에 그리기

#cv2.waitKey()							  # 이미지 위 키보드의 키 입력 대기 : 무한대기
cv2.waitKey(3000)						  # ms단위 => 1초 1000ms : 지정된 시간 후 자동 닫기
cv2.destroyAllWindows()					  # 키보드 입력 시 창 닫기기
```

- 그레이스케일(GrayScale) 이미지 로딩 및 출력 <hr>

```python
## 이미지 로딩
grayImg = cv2.imread(FILE, cv2.IMREAD_GRAYSCALE)

## 이미지 정보 확인
print(f'형태 : {grayImg.shape}')
print(f'차원 : {grayImg.ndim}D')
print(f'타입 : {grayImg.dtype}')
```

```python
## 이미지 출력
plt.imshow(grayImg, cmap='gray')
plt.title('Strawberry')
plt.show()
```
